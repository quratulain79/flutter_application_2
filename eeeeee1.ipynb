```python
!pip install -q tensorflow==2.15.0 protobuf==3.20.* >/dev/null

import os, cv2, random, shutil, glob, json, math, itertools
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, MobileNetV2
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# --- 1. CONFIGURATION & PATHS ---
random.seed(42); np.random.seed(42); tf.random.set_seed(42)
print(f"TensorFlow Version: {tf.__version__}")

IMG_SIZE = (224, 224)
SEG_SIZE = (128, 128) # Smaller size for U-Net speed
BATCH_SIZE = 32
TARGET_COUNT = 1000  # Har class ki 1000 images karega (adjust kar lo agar time kam hai)

DATA_DIR = "/kaggle/input/the-iqothnccd-lung-cancer-dataset/The IQ-OTHNCCD lung cancer dataset"
# Kaggle dataset mein kabhi kabhi Invalid Class path alag hota hai
INVALID_DIR_OPTIONAL = "/kaggle/input/invalid-class-dataset/Invalid Class"

# Working Directories Structure
BASE_DIR = "/kaggle/working/"
AUG_DIR = os.path.join(BASE_DIR, "1_augmented_raw")       # Step 1: Raw Augmented Data
MASKED_DIR = os.path.join(BASE_DIR, "2_masked_clean")     # Step 2: Background Removed Data
SPLIT_DIR = os.path.join(BASE_DIR, "3_splitted_data")     # Step 3: Train/Val/Test
RESULTS_DIR = os.path.join(BASE_DIR, "results")
TFLITE_DIR = os.path.join(RESULTS_DIR, "tflite_models")

for d in [AUG_DIR, MASKED_DIR, SPLIT_DIR, RESULTS_DIR, TFLITE_DIR]:
    os.makedirs(d, exist_ok=True)

classes = ["Normal cases", "Bengin cases", "Malignant cases", "Invalid Class"]

# --- 2. DATA AUGMENTATION (RAW IMAGES) ---
print("\n=== STEP 1: DATA AUGMENTATION (RAW) ===")

aug_datagen = ImageDataGenerator(
    rotation_range=20, width_shift_range=0.1, height_shift_range=0.1,
    zoom_range=0.1, horizontal_flip=True, fill_mode='nearest'
)

def augment_class(cls, class_source_path):
    if not os.path.exists(class_source_path): return
    os.makedirs(os.path.join(AUG_DIR, cls), exist_ok=True)
    
    images = [f for f in os.listdir(class_source_path) if f.lower().endswith(('.png','.jpg','.jpeg'))]
    print(f"[{cls}] Found {len(images)} originals. Augmenting to {TARGET_COUNT}...")
    
    # Copy originals first
    count = 0
    for img_name in images:
        try:
            shutil.copy(os.path.join(class_source_path, img_name), os.path.join(AUG_DIR, cls, f"org_{img_name}"))
            count += 1
        except: pass
        
    # Augment loop
    while count < TARGET_COUNT and len(images) > 0:
        for img_name in images:
            if count >= TARGET_COUNT: break
            img = cv2.imread(os.path.join(class_source_path, img_name))
            if img is None: continue
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, IMG_SIZE)
            x = np.expand_dims(img, 0)
            
            aug_iter = aug_datagen.flow(x, batch_size=1)
            aug_img = next(aug_iter)[0].astype(np.uint8)
            
            save_path = os.path.join(AUG_DIR, cls, f"aug_{count}.jpg")
            cv2.imwrite(save_path, cv2.cvtColor(aug_img, cv2.COLOR_RGB2BGR))
            count += 1
    print(f"[{cls}] Done. Total: {len(os.listdir(os.path.join(AUG_DIR, cls)))}")

# Run Augmentation Logic
for cls in classes:
    # Logic to find the folder
    src = os.path.join(DATA_DIR, cls)
    if cls == "Invalid Class" and not os.path.exists(src):
        if os.path.exists(INVALID_DIR_OPTIONAL):
            src = INVALID_DIR_OPTIONAL
    
    if os.path.exists(src):
        augment_class(cls, src)
    else:
        print(f"Skipping {cls}, source folder not found.")

# --- 3. TRAIN SEGMENTATION MODEL (U-NET) ---
print("\n=== STEP 2: TRAIN SEGMENTATION MODEL (U-NET) ===")
# We only train on Valid Lungs (Normal/Benign/Malignant). Invalid is excluded.

def create_otsu_mask(img, target_size=SEG_SIZE):
    # Auto-generate mask using Otsu Thresholding
    gray = cv2.cvtColor((img * 255).astype(np.uint8), cv2.COLOR_RGB2GRAY)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    gray = clahe.apply(gray)
    gray = cv2.resize(gray, target_size)
    _, mask = cv2.threshold(gray, 0, 1, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    kernel = np.ones((3, 3), np.uint8)
    mask = cv2.morphologyEx(mask.astype(np.uint8), cv2.MORPH_CLOSE, kernel)
    return mask[..., np.newaxis].astype(np.float32)

X_seg, Y_seg = [], []
seg_train_classes = ["Normal cases", "Bengin cases", "Malignant cases"] 

for cls in seg_train_classes:
    cls_path = os.path.join(AUG_DIR, cls)
    if not os.path.exists(cls_path): continue
    files = os.listdir(cls_path)
    random.shuffle(files)
    # Taking 400 images per class for Seg training to be fast
    for f in files[:400]: 
        img = cv2.imread(os.path.join(cls_path, f))
        if img is None: continue
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, SEG_SIZE).astype(np.float32) / 255.0
        mask = create_otsu_mask(img)
        X_seg.append(img)
        Y_seg.append(mask)

X_seg = np.array(X_seg); Y_seg = np.array(Y_seg)
print(f"Segmentation Training Data Shape: {X_seg.shape}")

# Build U-Net Architecture
def build_unet(input_shape):
    inputs = layers.Input(input_shape)
    c1 = layers.Conv2D(32, 3, activation='relu', padding='same')(inputs); p1 = layers.MaxPooling2D()(c1)
    c2 = layers.Conv2D(64, 3, activation='relu', padding='same')(p1); p2 = layers.MaxPooling2D()(c2)
    c3 = layers.Conv2D(128, 3, activation='relu', padding='same')(p2)
    u2 = layers.Conv2DTranspose(64, 2, strides=2, padding='same')(c3); u2 = layers.concatenate([u2, c2])
    c4 = layers.Conv2D(64, 3, activation='relu', padding='same')(u2)
    u1 = layers.Conv2DTranspose(32, 2, strides=2, padding='same')(c4); u1 = layers.concatenate([u1, c1])
    c5 = layers.Conv2D(32, 3, activation='relu', padding='same')(u1)
    outputs = layers.Conv2D(1, 1, activation='sigmoid')(c5)
    return models.Model(inputs, outputs)

seg_model = build_unet((SEG_SIZE[0], SEG_SIZE[1], 3))
seg_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print("Training U-Net...")
seg_model.fit(X_seg, Y_seg, batch_size=32, epochs=12, verbose=1) # 12 Epochs
seg_model.save(os.path.join(RESULTS_DIR, "seg_model.h5"))

# --- 4. APPLY MASKING (CREATE CLEAN DATASET) ---
print("\n=== STEP 3: MASKING & CLEANING DATASET ===")
# This step creates the Black Background images.
# If class is "Invalid", U-Net detects nothing -> Mask is Black -> Image becomes Black.

for cls in classes:
    os.makedirs(os.path.join(MASKED_DIR, cls), exist_ok=True)
    src_path = os.path.join(AUG_DIR, cls)
    if not os.path.exists(src_path): continue
    
    files = os.listdir(src_path)
    print(f"Processing Masking for {cls}...")
    
    for f in files:
        img_path = os.path.join(src_path, f)
        original = cv2.imread(img_path)
        if original is None: continue
        original_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
        
        # Predict Mask
        inp = cv2.resize(original_rgb, SEG_SIZE).astype(np.float32) / 255.0
        mask_pred = seg_model.predict(np.expand_dims(inp, 0), verbose=0)[0]
        
        # Threshold (Clean mask)
        mask_bin = (mask_pred > 0.5).astype(np.uint8)
        
        # Resize mask to original
        mask_full = cv2.resize(mask_bin, (original.shape[1], original.shape[0]))
        
        # Apply Mask (Black Background)
        mask_3ch = np.stack((mask_full,)*3, axis=-1)
        masked_img = original_rgb * mask_3ch
        
        # Save
        cv2.imwrite(os.path.join(MASKED_DIR, cls, f), cv2.cvtColor(masked_img.astype(np.uint8), cv2.COLOR_RGB2BGR))

# --- 5. SPLIT DATA (TRAIN/VAL/TEST) ---
print("\n=== STEP 4: SPLITTING DATA ===")
for s in ["train", "val", "test"]:
    for cls in classes:
        os.makedirs(os.path.join(SPLIT_DIR, s, cls), exist_ok=True)

for cls in classes:
    src = os.path.join(MASKED_DIR, cls)
    if not os.path.exists(src): continue
    files = os.listdir(src)
    random.shuffle(files)
    n = len(files)
    train_n = int(n*0.7)
    val_n = int(n*0.15)
    
    for f in files[:train_n]: shutil.copy(os.path.join(src, f), os.path.join(SPLIT_DIR, "train", cls, f))
    for f in files[train_n:train_n+val_n]: shutil.copy(os.path.join(src, f), os.path.join(SPLIT_DIR, "val", cls, f))
    for f in files[train_n+val_n:]: shutil.copy(os.path.join(src, f), os.path.join(SPLIT_DIR, "test", cls, f))

# --- 6. TRAIN CLASSIFIER (MOBILENET) ---
print("\n=== STEP 5: TRAIN CLASSIFIER (MobileNetV2) ===")

train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
val_test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_directory(os.path.join(SPLIT_DIR, "train"), target_size=IMG_SIZE, batch_size=BATCH_SIZE)
val_gen = val_test_datagen.flow_from_directory(os.path.join(SPLIT_DIR, "val"), target_size=IMG_SIZE, batch_size=BATCH_SIZE)
test_gen = val_test_datagen.flow_from_directory(os.path.join(SPLIT_DIR, "test"), target_size=IMG_SIZE, batch_size=1, shuffle=False)

# MobileNetV2
base = MobileNetV2(input_shape=IMG_SIZE+(3,), include_top=False, weights='imagenet')
base.trainable = False
x = layers.GlobalAveragePooling2D()(base.output)
x = layers.Dropout(0.3)(x)
out = layers.Dense(len(train_gen.class_indices), activation='softmax')(x)

clf_model = models.Model(base.input, out)
clf_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Class Weights (To handle imbalance if any)
labels = train_gen.classes
class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)
class_weights_dict = dict(enumerate(class_weights))

history_clf = clf_model.fit(train_gen, validation_data=val_gen, epochs=8, class_weight=class_weights_dict)
clf_model.save(os.path.join(RESULTS_DIR, "clf_model.h5"))

# --- 7. EVALUATION ---
print("\n=== STEP 6: EVALUATION ===")

y_true = test_gen.classes
y_pred_probs = clf_model.predict(test_gen, verbose=1)
y_pred = np.argmax(y_pred_probs, axis=1)

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=list(test_gen.class_indices.keys())))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=list(test_gen.class_indices.keys()), yticklabels=list(test_gen.class_indices.keys()))
plt.title("Confusion Matrix")
plt.savefig(os.path.join(RESULTS_DIR, "confusion_matrix.png"))
plt.show()

# --- 8. DETAILED VISUALIZATION (ONE PER CLASS) ---
print("\n=== STEP 7: PIPELINE VISUALIZATION (One per Class) ===")

def visualize_each_class_example():
    inv_map = {v: k for k, v in test_gen.class_indices.items()}
    
    # Loop through every class
    for target_cls in classes:
        if target_cls not in test_gen.class_indices: continue
        
        # Find indices of this class
        cls_idx = test_gen.class_indices[target_cls]
        indices = np.where(test_gen.classes == cls_idx)[0]
        
        if len(indices) == 0: continue
            
        # Pick a random image from this class
        idx = np.random.choice(indices)
        fname = test_gen.filenames[idx] # e.g., "Malignant cases/aug_50.jpg"
        
        # Important: Load the RAW image from AUG_DIR (trace back)
        raw_path = os.path.join(AUG_DIR, fname)
        if not os.path.exists(raw_path): continue
            
        original = cv2.imread(raw_path)
        original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
        
        # 1. SEGMENTATION
        seg_in = cv2.resize(original, SEG_SIZE).astype(np.float32) / 255.0
        mask_pred = seg_model.predict(np.expand_dims(seg_in, 0), verbose=0)[0,:,:,0]
        mask_full = cv2.resize((mask_pred > 0.5).astype(np.uint8), (original.shape[1], original.shape[0]))
        
        # 2. MASKING (Cleaning)
        mask_3ch = np.stack((mask_full,)*3, axis=-1)
        clean_img = (original * mask_3ch).astype(np.uint8)
        
        # 3. CLASSIFICATION
        clf_in = cv2.resize(clean_img, IMG_SIZE)
        clf_in = preprocess_input(clf_in.astype(np.float32))
        probs = clf_model.predict(np.expand_dims(clf_in, 0), verbose=0)[0]
        pred_idx = np.argmax(probs)
        pred_label = inv_map[pred_idx]
        conf = probs[pred_idx] * 100
        
        # 4. VISUAL EXTRAS
        heatmap = cv2.applyColorMap(np.uint8(255 * mask_full), cv2.COLORMAP_JET)
        heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)
        overlay = cv2.addWeighted(original, 0.6, cv2.resize(heatmap, (original.shape[1], original.shape[0])), 0.4, 0)
        
        bbox_img = original.copy()
        contours, _ = cv2.findContours(mask_full, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        for cnt in contours:
            x,y,w,h = cv2.boundingRect(cnt)
            if cv2.contourArea(cnt) > 50:
                cv2.rectangle(bbox_img, (x,y), (x+w, y+h), (0,255,255), 2)
        
        # Severity
        pct = (mask_pred > 0.5).mean() * 100
        sev = "None"
        if pct > 0:
            if pct < 15: sev = "Mild"
            elif pct < 30: sev = "Moderate"
            else: sev = "Severe"

        # PLOTTING
        plt.figure(figsize=(20, 4))
        plt.suptitle(f"Target Class: {target_cls}", fontsize=14, fontweight='bold')
        
        plt.subplot(1, 5, 1); plt.imshow(original); plt.title("1. Original Raw")
        plt.axis('off')
        
        plt.subplot(1, 5, 2); plt.imshow(mask_full, cmap='gray'); plt.title("2. Seg Mask")
        plt.axis('off')
        
        plt.subplot(1, 5, 3); plt.imshow(clean_img); plt.title("3. Cleaned (Input)")
        plt.axis('off')
        
        plt.subplot(1, 5, 4); plt.imshow(overlay); plt.title(f"4. Heatmap ({sev})")
        plt.axis('off')
        
        col = 'green' if pred_label == target_cls else 'red'
        plt.subplot(1, 5, 5); plt.imshow(bbox_img)
        plt.title(f"5. Pred: {pred_label}\nConf: {conf:.1f}%", color=col, fontweight='bold')
        plt.axis('off')
        
        plt.show()

visualize_each_class_example()

# --- 9. TFLITE CONVERSION ---
print("\n=== STEP 8: EXPORT TFLITE ===")
converter = tf.lite.TFLiteConverter.from_keras_model(clf_model)
with open(os.path.join(TFLITE_DIR, "clf_model.tflite"), "wb") as f: f.write(converter.convert())

converter = tf.lite.TFLiteConverter.from_keras_model(seg_model)
with open(os.path.join(TFLITE_DIR, "seg_model.tflite"), "wb") as f: f.write(converter.convert())

print(f"âœ… Pipeline Finished Successfully. Models saved in: {TFLITE_DIR}")
```
